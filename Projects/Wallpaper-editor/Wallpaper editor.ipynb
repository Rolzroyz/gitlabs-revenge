Wallpaper editor
===
A program disgned to merge two images together for a dual moniter setup
---
Note: This program assumes you are using images the same size as your monitors

```python
from IPython import display #to display
import PIL.Image as Image  #to edit
filename_left ="all-black.png"      #image you would like on your left monitor
filename_right = "firefox-logo.png" #image you would like on your right monitor #firefox->mozzila
wallpaper_name = "new-wallpaper.png"
display.Image(filename=filename_left)
```

```python
display.Image(filename=filename_right)
```

```python
img1 = Image.open(filename_left) #open the images
img2 = Image.open(filename_right)

img1_width = img1.size[0]        #get the size parameters
img2_width = img2.size[0]
total_width = img1_width + img2_width #compute width

img1_height = img1.size[1]
img2_height = img2.size[1]
if img1_height > img2_height:     #choose taller image as height
    total_height = img1_height
else:
    total_height = img2_height

new_wallpaper = Image.new('RGB', (total_width,img1_height)) #creat image
new_wallpaper.paste(img1, (0,0))  #past in images
new_wallpaper.paste(img2, (img1_width,0))
new_wallpaper.save(wallpaper_name) #save it
display.Image(wallpaper_name)      #check that it worked
```
